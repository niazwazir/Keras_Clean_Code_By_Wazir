```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, BatchNormalization, LeakyReLU,add
```

```python
def Wazir_Res_block():
  _input = Input(shape=(None, None, 3))
  conv1 = Conv2D(filters=64, kernel_size=(3, 3), strides=(1, 1), padding='same', activation='relu')(_input)
  BN = BatchNormalization()(conv1)
  LR = LeakyReLU(alpha=0.2)(BN)
  conv2 = Conv2D(filters=3, kernel_size=(3, 3), strides=(1, 1), padding='same', activation='linear')(LR)
  BN1 = BatchNormalization()(conv2)
  out = add(inputs=[_input, BN1])
  model = Model(inputs=_input, outputs=out)
  return model

model = Wazir_Res_block()
```

```python
tf.keras.utils.plot_model(
    model,to_file='model.png',show_shapes=False,show_layer_names=True,rankdir='LR',expand_nested=False,dpi=96)
```
